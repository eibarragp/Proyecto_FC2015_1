```julia
using PyPlot
```

```julia
ϵ_o=8.85e-12;
μ_o=(4e-7)*π;
K=sqrt(ϵ_o/μ_o);
```

###### Lista de índices de refracción

```julia
naire=1.0
nvidrio=1.5
```

```julia
function snell(θ_i,n_i,n_t)
    θ_t=asin((n_i/n_t)*sin(θ_i))
    θ_t
end
```

$$\delta= 2\pi \frac{d}{\lambda} n_f \cos(\theta_{t_f})$$

Sea $\rho =\frac{d}{\lambda}$

$$\delta= 2\pi \rho n_f \cos(\theta_{t_f})$$

```julia
function fase(θ_i,n_i,n_f,ρ)
    δ=2*pi*ρ*n_f*cos(snell(θ_i,n_i,n_f))
    δ
end
```

```julia
function Mt(θ_i,n_i,n_f,ρ)
    δ=fase(θ_i,n_i,n_f,ρ)
    Yf=K*n_f*cos(snell(θ_i,n_i,n_f))
    #if modo=="TE"   
    #elseif modo=="TM"
    #    Yf=K*n_f/cos(snell(θ_i,n_i,n_f))
    #end
    Mt11=cos(δ)
    Mt12=(im*sin(δ))/Yf
    Mt21=im*sin(δ)*Yf
    Mt22=cos(δ)
    
    [Mt11 Mt12; Mt21 Mt22]
end
```

```julia
function RT(M,n_o,n_s,n_f,θ_i)
    Yo=K*n_o*cos(θ_i)
    Ys=K*n_s*cos(snell(snell(θ_i,n_o,n_f),n_f,n_s))
    
    numerador_r= Yo*M[1,1] + Yo*Ys*M[1,2] - M[2,1] - Ys*M[2,2]
    numerador_t= 2*Yo
    denominador= Yo*M[1,1] + Yo*Ys*M[1,2] + M[2,1] + Ys*M[2,2]
    
    r=numerador_r/denominador
    t=numerador_t/denominador
    
    R=(norm(r))^2
    T=(n_s/n_o)*(cos(snell(snell(θ_i,n_o,n_f),n_f,n_s))/cos(θ_i))*(norm(t))^2
    #T=1-R
    
    [R,T]
end
```

```julia
Valores=Vector{Float64}[]
thetas=[0:0.01:π/2]
for θ in thetas
    PlacaVidrio=Mt(θ,naire,nvidrio,2)
    Coef=RT(PlacaVidrio,naire,nvidrio,nvidrio,θ)
    push!(Valores,Coef)
end
angulos=[θ*180/π for θ in thetas]
Valores=hcat(Valores...)';
figure(figsize=(6,4))
plot(angulos,Valores[:,1])
plot(angulos,Valores[:,2])
xlabel(L"$\theta_i$")
ylabel(L"$R,T$")
legend(["Reflectancia","Transmitancia"],bbox_to_anchor=(1.60, 0.55))
title("Interfaz aire-vidrio, modo TE")
```

```julia
figure(figsize=(6,4))
for j in [1.5,1.7,2.0,3.0]
    n2=j
    Valores=Vector{Float64}[]
    rhos=[0:0.001:1]
    for ρ in rhos
        Placa=Mt(0,naire,n2,ρ/n2)
        Coef=RT(Placa,naire,nvidrio,n2,0)
        push!(Valores,Coef)
    end
    Valores=hcat(Valores...)';
    plot(rhos,Valores[:,1])
    xlabel(L"$\rho/n_f$")
    ylabel(L"$R,T$")
    title("Interfaz aire-vidrio, modo TE")
    #vlines([0.25,0.5,0.75],0,1)
end
```
